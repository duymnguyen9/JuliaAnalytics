```julia
#writing a fun julia guessing number game
function guessingame()
    right_answer = 56
    println("Please Enter a number: ")
    count = 1
    guess = 0
    try
        guess = parse(Int,readline())
    catch y
        println("Not a number!")
        
    end
    while guess != right_answer
        if guess > right_answer
            println("Your guess is too high, try again")
            elseif guess < right_answer
                println("Your guess is too low, try again")
        end
        count += 1
        guess = parse(Int,readline())
    end
    println("your answer is correct, you inputted: ", guess, " Number of guess: ", count)
end
```

```julia
guessingame()
```
